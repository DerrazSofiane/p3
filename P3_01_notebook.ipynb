## Définition de fonctions utiles

```python
def top_n(df: object, columns: [str, list], n:int, normalize=False) -> "pandas.Series":
    """Donne le top n d'une variable (colonne) donnée.
    
    Args:
      df:
        Nom de la dataframe.
      columns:
        Nom de(s) colonne(s). Peut être un string ou une liste.
      n:
        Le nombre de ligne que l'on souhaite avoir.
      normalize:
        Argument de la méthode value_counts(), si l'on veut une colonne normalisée.
        
    Retourne:
      Une série pandas avec le top n pour une colonne donnée.
      Affiche les résultats dans l'ordre croissant.
      
      Par exemple:
        >>> top_n(data, 'countries_fr', 10)
        retourne le top 10 des pays les plus présents.
    """
    top10 = df[columns].value_counts(normalize=normalize)
    
    return top10[:n][::-1].round(2)


def plot_hist(df: object,
              variable: object,
              x_axis: str,
              color='royalblue',
              title='Title') -> "sns.barplot":
    """Affiche un histograme en barre horizontal pour une variable donnée
    
    Args:
      df:
        Dataframe à analyser.
      variable:
        Series pandas correspondant au count (axe x).
      x_axis:
        Nom de la variable.
      color:
        Couleur des barres.
        
    Retourne:
      La variable contenant le graphique.
    
      Par exemple:
        plot_hist(top_15_countries, top_15_countries['count'], 'count')
    """
    plt.figure(figsize=(12, 8))
    plt.title(title)
    ax = sns.barplot(x=x_axis, y=df.index, data=df, color=color)

    # Affiche la valeur à droite de chaque barre
    for index, value in enumerate(variable):
        ax.text(value, index, str(value)+"%", color='black', fontweight='normal')
        
    return ax
```

##### Fonction de nettoyage

```python
def nettoyage(df):
    """Nettoie la dataframe en effectuant plusieurs étapes:
    1) Suppression des lignes sans données
    2) Suppression des colonnes sans données
    3) Suppression des valeurs doubles
    
    Retourne la dataframe nettoyée.
    """
    nrows, ncols = data.shape

    # Supprimer les lignes sans données
    data.dropna(how='all', inplace=True)

    # Supprimer les colonnes sans données
    empty_variables = data.columns[data.isnull().sum() == data.shape[0]]
    data.dropna(how='all', axis=1, inplace=True)

    # Supprimer les valeurs en double
    duplicates_count = data.duplicated().sum()
    data.drop_duplicates(inplace=True)

    print(f"Nombre de lignes vides supprimées: {nrows - data.shape[0]}")
    print(f'Nombre de colonnes vides supprimées: {len(empty_variables)}')
    print(f'Colonnes supprimées:\n{empty_variables}')
    print(f'Nombre de doublons supprimés: {duplicates_count}')
    print(f'\nDimensions du jeu de données: {data.shape}')
    
    return df

data = nettoyage(data)
data
```

#### L'énergie

Maintenant que nous avons supprimer les valeurs abérrantes pour le gras, le sucre et les protéines, on peut maintenant traiter les valeurs abérrantes pour l'énergie grâce à la formule suivante :

> Énergie [kcal] = 9 * Lipides [g] + 4 * Glucides [g] + 4 * Protéines [g] (9/4/4).

Source : https://thermes-barbotan.fr/wafx_res/Files/Calcul_20Ration_20Energ_C3_A9tique.pdf

Comme l'énergie est en kJ dans le jeu de données original, nous allons convertir le résultat de cette formule en kJ. La conversion de kcal vers kJ s'effectue en multipliant l'énergie (en kcal) par 4,184.

> Énergie [kJ] = 4,184 * Énergie [kcal]

```python
data["energy_100g"].describe()
```

```python
plt.xlim(0, 100000)
ax = sns.distplot(data['energy_100g'], fit=stats.norm, kde=False)
ax.set_title("Distribution par rapport à la loi normale d'energy_100g")
plt.show()
```

On peut observer une distribution qui ne suit pas du tout la loi normale, avec des valeurs sur l'axe des abcisses totalement abérrantes. On a pour valeur max 3000000 kJ, ce qui est impossible.

D'après cette source, la valeur max pour l'energie est de 900kcal/100g, pour les aliments les plus énergétiques. https://sante.journaldesfemmes.fr/calories/classement/aliments/calories.
Donc <code>900kcal*4,184 = 3766kJ</code>

Après correction des abérrations pour l'énergie, on peut s'attendre à des valeurs maximales aux alentours de 4000.

```python
# Calcul de l'énergie (kcal)
data["energy_100g"] = 4 * data.carbohydrates_100g + 9 * data.fat_100g + 4 * data.proteins_100g

# Conversion en kJ
data["energy_100g"] = data["energy_100g"] * 4.184
data["energy_100g"].describe()
```

On dispose d'une valeur max de 4572, qui est assez elevé mais qui reste beaucoup plus cohérent que 3e^6kJ.

```python
ax = sns.distplot(data['energy_100g'], fit=stats.norm, kde=False)
ax.set_title("Distribution par rapport à la loi normale d'energy_100g")
plt.show()
```

La courbe a l'air de suivre la loi normale, ce qui montre que le traitement des abérration a été suffisamment pertinent.

Certaines lignes regroupent plusieurs fois le même allergène. Le lait existe en doublons mais avec une majuscule. nous pouvons les rassembler en un seul allergène.

```python
top_10_allergen_index = list(top_10_allergen.index)

# Conversion des majuscules en minuscle
top_10_allergen_index = [each_value.lower() for each_value in top_10_allergen_index]

for index, value in enumerate(top_10_allergen_index):
    if value == "blé, blé":
        top_10_allergen_index[index] = "blé"
    elif value == "lait, lait":
        top_10_allergen_index[index] = "lait"
    elif value == "lactose":
        top_10_allergen_index[index] = "lait"

top_10_allergen.index = top_10_allergen_index
top_10_allergen = top_10_allergen.groupby(level=0).sum()
top_10_allergen
```

# ACP

https://www.analyticsvidhya.com/blog/2016/03/pca-practical-guide-principal-component-analysis-python/

```python
from sklearn import decomposition, preprocessing


def make_acp(list_of_cols):
    # choix du nombre de composantes à calculer
    n_comp = len(list_of_cols)
    print(f"{n_comp} composantes")
    
    # selection des colonnes à prendre en compte dans l'ACP
    data_pca = data_fr[list_of_cols].fillna(0)
    X = data_pca.values
    features = data_pca.columns
    
    # affichage de la matrice de corrélation
    d = pd.DataFrame(data=data_fr, columns=list_of_cols)
    corr = d.corr()
    mask = np.triu(np.ones_like(corr, dtype=bool))
    f, ax = plt.subplots(figsize=(11, 9))
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.8, center=0,
    square=True, linewidths=.5, cbar_kws={"shrink": .5})
    plt.show()
    
    # Centrage et Réduction
    std_scale = preprocessing.StandardScaler().fit(X)
    X_scaled = std_scale.transform(X)

    # Calcul des composantes principales
    pca = decomposition.PCA(n_components=n_comp)
    pca.fit(X_scaled)

    # Eboulis des valeurs propres
    display_scree_plot(pca)

    # Cercle des corrélations
    pcs = pca.components_
    display_circles(pcs, n_comp, pca, [(0,1)],
                    labels = np.array(features))
    
    plt.show()
```

```python
test_cols = ["energy_100g","fat_100g","carbohydrates_100g", "fiber_100g","proteins_100g","sodium_100g", "saturated_fat_100g",
             "trans_fat_100g", "cholesterol_100g", "sugars_100g", "salt_100g"]
make_acp(test_cols)
```

Comme le tableau dispose de plusieurs dimensions, il est très difficile d'observer la corrélation d'une variable par rapport à tout le jeu de données. Les données sont donc représentées par les composantes principales F1 et F2. À l’intérieur du cercle, il y a des flèches qui partent du centre. Elles sont plus ou moins grandes, et si elles se dirigent vers une même direction cela signifie que ces variables sont corrélées.

On observe de nombreuses paires de variables qui sont corrélées :
* salt/sodium
* fat/saturated_fat
* carbohydrates/sugars

L'énergie a également l'air d'être très corrélé au gras.

Il serait alors possible, à partir de cette observation, de synthétiser ces paires de variable en une seule variable.

# ANOVA

https://www.analyticsvidhya.com/blog/2020/06/introduction-anova-statistics-data-science-covid-python/

Afin de vérifier si la pnns_groups_2 influence rééllement la proportion moyenne de gras, nous pouvons réaliser une ANOVA (Analyse de la variance). Le choix de ce test est dû au fait que nous étudions 1 variable qualitative comparativement à une variable quantitative.

Les hypothèses posées seront donc les suivantes :

* H0 : La proportion de gras est similaire pour chaque groupe (et donc la catégorie n'a aucune influence sur la proportion de gras).
* H1 : Une ou plusieurs proportions sont inégales.

```python
food_group_mean = data_fr.groupby('pnns_groups_1').mean()
moyenne_gras_par_groupe = food_group_mean[["fat_100g"]].sort_values("fat_100g", ascending=False)
moyenne_gras_par_groupe
```

```python
moyenne_gras_par_groupe.describe()
```

Pour commencer, nous pouvons projeter les  boxplots de la répartition moyenne de gras par catégorie pnns_groups_1.

```python
fig = plt.figure(figsize=(12,8))
ax = sns.boxplot(data=data_fr, x='pnns_groups_1', y='fat_100g')
plt.setp(ax.get_xticklabels(), rotation=45)
plt.title("Groupe PNNS1 sur le gras au 100g")
plt.show()
```

Les boîtes ont l'air d'être inégalement répartit sur le graphiques, bien que certains groupes d'aliment ont l'air d'avoir une répartition de gras similaire.

https://gist.github.com/sinhrks/9579355

```python
fig = plt.figure(figsize=(12,8))
fig.suptitle("Proportion moyenne de gras par groupe de produit PNNS1", fontsize=20)

values = ['{0:0.2f}%'.format(s) for s in moyenne_gras_par_groupe["fat_100g"]]
squarify.plot(sizes=moyenne_gras_par_groupe['fat_100g'], label=moyenne_gras_par_groupe.index, value=values, alpha=.8)
plt.axis('off')
plt.show()
```

Nous pouvons, à partir de cette visualisation, en déduire que le groupe d'aliment a une influence sur la proportion de gras au 100g.

Afin d'en être certain, nous pouvons quantifier cette influence grâce à la p value, permettant de valider ou non les hypothèses.

https://www.statsmodels.org/stable/generated/statsmodels.stats.anova.anova_lm.html

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

anova_category = smf.ols('fat_100g~pnns_groups_1', data=data_fr).fit()
anova_result = sm.stats.anova_lm(anova_category, typ=2)
anova_result
```

Les résultats du test de Fisher nous indiquent ici une p-value de 0 pour l'ensemble des catégories, donc inferieur au niveau de test de 5%. Nous rejettons donc l'hypothèse H0 selon laquelle la proportion de gras est similaire pour chaque groupe.
La catégorie de produit a donc bien une influence sur la proportion de gras au 100g.
